```python
def correlation_matrix(df):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm
    fig = plt.figure(figsize=(16,12))
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet',30)
    cax = ax1.imshow(df.corr(),interpolation="nearest" ,cmap = cmap)
    ax1.grid(True)
    plt.title('Data set feature correlation\n',fontsize=15)
    labels=df.columns
    ax1.set_xticklabels(labels,fontsize=12)
    ax1.set_yticklabels(labels,fontsize=12)
    #fig.colorbar(cax,ticks=[0.1*i for i in range(-11,11)])
    plt.show()

    
correlation_matrix(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x = df.drop('radius_mean',axis=1)
y = df['radius_mean']
```

```python
x = scaler.fit_transform(x)
```

```python
dfnew = pd.DataFrame(data=x,columns=df.columns[1:])
```

```python
dfnew.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
dfnew_pca = pca.fit(dfnew)
```

```python
plt.figure(figsize=(9,5))
plt.scatter(x=[i+1 for i in range(len(dfnew_pca.explained_variance_ratio_))],
y = dfnew_pca.explained_variance_ratio_,
s=200,alpha=0.75,c='red',edgecolor='m')
plt.grid(True)
plt.xticks([i+1 for i in range(len(dfnew_pca.explained_variance_ratio_))])
plt.yticks(fontsize=15)
plt.show()
```

```python
dfx_trans = pca.transform(dfnew)
dfx_trans = pd.DataFrame(data=dfx_trans)
dfx_trans.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(dfx_trans[0],dfx_trans[1],c=df['radius_mean'],edgecolors='m',alpha=0.75,s=150)
plt.grid(True)
plt.title("Class seperation")
plt.xlabel("PC-1",fontsize = 13)
plt.ylabel("PC-2",fontsize = 13)
plt.show()
```
